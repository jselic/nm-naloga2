# NM DN2: Besselova Funkcija
Naloga je izračunati integral $\frac{1}{\pi} \int_{0}^{\pi} \left (\cos(x \sin(t)) \right) \, dt$.
Tega se bomo lotili s sestavljenim trapeznim pravilom, in rezultate primerjali z vgrajeno besselovo funkcijo ničtega reda iz knjižnice scipy. Najprej ročno preverimo na eni vrednosti:

```python
jv(0,1)
```

```python
#Implementacija se nahaja v datoteki src/Functions.py
bessel_0(1)
```

Poglejmo kako se obnese na tesni funkciji s 1000 naključnimi števili:

```python
def test_function(n=1000):
    gt = []
    computed = []
    for _ in range(n):
        rint = np.random.randint(0,200)
        gt.append(jv(0,rint))
        computed.append(bessel_0(rint))
    return gt, computed
```

```python
#Izvajanje lahko traja par sekund
results = test_function()
```

```python
#Pogledamo ali se rezultati razlikujejo za 10e-10 ali manj?
np.all(np.isclose(results[0],results[1],atol=1e-10))
```

```python
#Pogledamo še ročno
max([results[0][i]-results[1][i] for i,_ in enumerate(results[0])])
```

Največja razlika je probližno 10e-15

# DN2: Ploščina Bézierove krivulje
Naloga je izračunati ploščino Bezierove, krivulje. Ki jo določajo podane kontrolne točke. Za to bomo uporabili De Casteljau-jev algoritem da iz kontrolnih točk generiramo $n$ dejanskih. Nato bomo poiskali točko, kjer krivulja preseka samo sebe - in odstranili njene 'repe'. Ploščino dobljenega poligona bomo nato zračunali z algoritmom 'Shoelace' (Gauss' area formula)

```python
#Kontrolne točke
control_points = np.array([
    (0, 0), (1, 1), (2, 3), (1, 4),
    (0, 4), (-1, 3), (0, 1), (1, 0)
])
```

Najprej poglejmo kako krivulja sploh zgleda.
Za to moramo najprej izračunati dejanske točke

```python
from src.Functions import point_from_bezier #De Casteljau
import matplotlib.pyplot as plt
```

Generiramo 1000 točk za vsak $t \in$ $[0,1]$, z korakom $\delta t = 0.001$

```python
t_values = np.linspace(0, 1, 1000)
points = [point_from_bezier(control_points, t) for t in t_values]
```

```python
#Izpišemo prvih 10 točk
points[:10]
```

```python
#In prikažemo dejansko krivuljo
plt.figure(figsize=(12,8))
points = np.array(points)
plt.plot(points[:, 0], points[:, 1], label='Bézier Curve (De Casteljau)')
plt.scatter(*zip(*control_points), color='red', label='Control Points')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Bézier Curve with Control Points (De Casteljau)')
plt.legend()
plt.grid()
plt.show()
```

Vidimo, da krivulja seka sama sebe pri približno $x=0.5$. Poiščimo tako točko

```python
from itertools import combinations
#ker je točk 1000, naje bo tudi toleranca 1/1000
close_pairs = close_pairs = [((i, j), points[i], points[j]) for (i, j) in combinations(range(len(points)), 2) if abs(points[i][0] - points[j][0]) < 1e-3 and abs(points[i][1] - points[j][1]) < 1e-3]
```

```python
close_pairs
```

```python
#Vzemimo samo del 
reduced_points = points[close_pairs[0][0][0]:close_pairs[0][0][1]]
```

Poglejmo kako zgleda krivulja brez "repov"

```python
plt.figure(figsize=(12,8))
reduced_points = np.array(reduced_points)
plt.plot(reduced_points[:, 0], reduced_points[:, 1], label='Bézier Curve (De Casteljau)')
plt.scatter(*zip(*control_points), color='red', label='Control Points')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Reduced Bézier Curve with Control Points (De Casteljau)')
plt.legend()
plt.grid()
plt.show()
```
